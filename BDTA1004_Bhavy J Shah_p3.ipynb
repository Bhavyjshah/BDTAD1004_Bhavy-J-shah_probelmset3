```python
import pandas as pd

# Step 2: Import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep='|')

# Step 4: Mean age per occupation
mean_age_per_occupation = users.groupby('occupation')['age'].mean()

# Step 5: Male ratio per occupation and sort
def male_ratio(series):
    male_count = (series == 'Male').sum()
    total = len(series)
    return male_count / total

male_ratio_per_occupation = users.groupby('occupation')['gender'].apply(male_ratio)
sorted_male_ratio = male_ratio_per_occupation.sort_values(ascending=False)

# Step 6: Minimum and maximum ages per occupation
min_max_ages_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])

# Step 7: Mean age per combination of occupation and sex
mean_age_per_combination = users.groupby(['occupation', 'gender'])['age'].mean()

# Step 8: Percentage of women and men per occupation
def gender_percentage(series):
    total = len(series)
    gender_counts = series.value_counts()
    return (gender_counts / total) * 100

gender_percentage_per_occupation = users.groupby('occupation')['gender'].apply(gender_percentage)

print("Mean age per occupation:\n", mean_age_per_occupation)
print("\nMale ratio per occupation and sorted:\n", sorted_male_ratio)
print("\nMinimum and maximum ages per occupation:\n", min_max_ages_per_occupation)
print("\nMean age per combination of occupation and sex:\n", mean_age_per_combination)
print("\nPercentage of women and men per occupation:\n", gender_percentage_per_occupation)
```

```python

import pandas as pd

# Step 2: Import the dataset
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url, sep='\t')

# Step 4: See the first 10 entries
print("First 10 entries:")
print(chipo.head(10))

# Step 5: Number of observations
num_observations = len(chipo)
print("\nNumber of observations:", num_observations)

# Step 6: Number of columns
num_columns = len(chipo.columns)
print("\nNumber of columns:", num_columns)

# Step 7: Print column names
print("\nColumn names:")
print(chipo.columns)

# Step 8: Index type
index_type = chipo.index
print("\nIndex type:", type(index_type))

# Step 9: Most-ordered item
most_ordered_item = chipo['item_name'].value_counts().idxmax()
print("\nMost-ordered item:", most_ordered_item)

# Step 10: Number of items ordered for most-ordered item
most_ordered_item_count = chipo[chipo['item_name'] == most_ordered_item]['quantity'].sum()
print("\nNumber of items ordered for most-ordered item:", most_ordered_item_count)

# Step 11: Most ordered item in choice_description column
most_ordered_choice = chipo['choice_description'].value_counts().idxmax()
print("\nMost ordered item in choice_description column:", most_ordered_choice)

# Step 12: Total items ordered
total_items_ordered = chipo['quantity'].sum()
print("\nTotal items ordered:", total_items_ordered)

# Step 13: Convert item price to float
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))

# Step 14: Revenue
revenue = (chipo['quantity'] * chipo['item_price']).sum()
print("\nRevenue for the period:", revenue)

# Step 15: Number of orders
num_orders = chipo['order_id'].nunique()
print("\nNumber of orders:", num_orders)

# Step 16: Average revenue per order
avg_revenue_per_order = revenue / num_orders
print("\nAverage revenue amount per order:", avg_revenue_per_order)

# Step 17: Number of different items sold
num_different_items = chipo['item_name'].nunique()
print("\nNumber of different items sold:", num_different_items)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the data from the provided CSV file
file_path = "C:\\Users\\bhavy\\Downloads\\arcade-revenue-vs-cs-doctorates.csv"
data = pd.read_csv(file_path)

# Step 2: Filter the data for the years 2000 to 2009
years = list(range(2000, 2010))
filtered_data = data[data['Year'].isin(years)]

# Step 3: Create a scatter plot
plt.figure(figsize=(10, 6))
for year in years:
    year_data = filtered_data[filtered_data['Year'] == year]
    plt.scatter(year_data['Total Arcade Revenue (billions)'], year_data['Computer Science Doctorates Awarded (US)'], label=str(year))

plt.title("Relationship Between Arcade Revenue and Computer Science PhDs (2000-2009)")
plt.xlabel("Total Arcade Revenue (billions)")
plt.ylabel("Computer Science Doctorates Awarded (US)")
plt.legend(title="Year")

plt.show()



```

```python
#Step:9
def stats(x):
    x = pd.Series(x)
    Min = x.min()
    Max = x.max()
    Mean = x.mean()
    Std = x.std()
    res = [Min,Max,Mean,Std]
    indx = ["Min","Max","Mean","Std"]
    res = pd.Series(res,index=indx)
    return res
#A different set of numbers for each location.
loc_stats = newData.apply(stats)
loc_stats
```
